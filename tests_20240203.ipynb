```python
def filter_active_spot_pairs(market_data):
    active_spot_pairs = {}
    for symbol, details in market_data.items():
        if details['active'] and 'SPOT' in details:
            active_spot_pairs[symbol] = details
    return active_spot_pairs

active_spot_pairs = filter_active_spot_pairs(market.market_data)
```
